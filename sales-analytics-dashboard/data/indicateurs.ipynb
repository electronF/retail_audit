```python
indicateur=pd.read_csv("Indicateurs & Formules(Feuil1).csv", sep=";")
# Supprime les colonnes qui contiennent UNIQUEMENT des NaN du DataFrame 'indicateur'
# Le résultat est stocké dans un nouveau DataFrame nommé 'indicateur_nettoye'
indicateur= indicateur.dropna(axis=1, how='all')
indicateur= indicateur.dropna(axis=0, how='all')

# Utilisation de la méthode rename()
indicateur= indicateur.rename(columns={'Unnamed: 0': 'Indicateurs'})
# Remplacez 'NomDeLaColonneASupprimer' par le nom exact affiché dans votre notebook.

# suppression d ela colonne nommée 'Unnamed: 4'
nom_colonne = 'Unnamed: 4' 

indicateur = indicateur.drop(columns=[nom_colonne])




# Afficher les nouvelles colonnes pour vérifier le changement


# Optionnel : Si vous voulez que la modification soit faite DIRECTEMENT sur 'indicateur' :
# indicateur.dropna(axis=1, how='all', inplace=True)

# Afficher les informations sur le nouveau DataFrame (colonnes restantes) pour vérifier
```

```python
with open("indicators_categories.txt", "r", encoding="utf-8") as file:
    categories=file.readlines()
    categories = [category.strip() for category in categories]
# exportation en fichier txt

categories
```

```python
# Dictionnaire des indicateurs par groupe

dictionnaire_final = {}
indicateur_courant = None # Clé du dictionnaire actuelle

for element in indicateur['Indicateurs']:
    if element in categories:
        # 1. On trouve un indicateur principal : il devient la nouvelle clé
        indicateur_courant = element
        dictionnaire_final[indicateur_courant] = []
    elif indicateur_courant is not None:
        # 2. On trouve une sous-catégorie : on l'ajoute à la liste de la clé courante
        dictionnaire_final[indicateur_courant].append(element)
    # else: L'élément est ignoré s'il n'est ni un indicateur principal, ni une sous-catégorie après un indicateur principal
    # Cette condition ne devrait normalement pas être nécessaire si votre colonne est propre.

# Affichage du résultat
import pprint
pprint.pprint(dictionnaire_final)
```
